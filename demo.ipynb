# 目標船艦周圍經緯度計算器 - Jupyter Notebook 示範

本 Notebook 示範如何匯入並使用 `ship_navigator` 套件，在目標船艦（中心點）的周圍每 10 度方位角，隨機生成 5-15 公里距離的經緯度點。

### 1. 環境設定與匯入套件
我們將 `src/` 目錄加入到系統路徑中，以便能夠直接載入 `ship_navigator` 模組。

```python
import sys
from pathlib import Path

# 動態將 src/ 目錄加入 sys.path
sys.path.insert(0, str(Path.cwd() / "src"))

# 匯入我們設計的類別與函數
from ship_navigator import Coordinate, generate_randomized_surrounding_points
```

### 2. 設定目標船艦中心經緯度並生成周圍點
我們設定中心點坐標，並呼叫函數計算 36 個方位的隨機距離座標。

```python
# 設定目標船艦中心點 (例如：緯度 22.0, 經度 120.0)
target_lat = 22.0
target_lon = 120.0
center = Coordinate(latitude=target_lat, longitude=target_lon)

print(f"目標船艦中心位置: 緯度 {center.latitude}, 經度 {center.longitude}")

# 生成周圍 36 個方位的座標點 (每 10 度，隨機 5-15 公里)
points = generate_randomized_surrounding_points(
    center_coord=center,
    bearing_step=10,
    min_dist_km=5.0,
    max_dist_km=15.0
)

print(f"成功生成 {len(points)} 個周圍經緯度座標點。")
```

### 3. 列印生成的經緯度座標

```python
# 格式化列印前 10 個生成的座標作為範例
print(f"{'方位角 (度)':<12} | {'隨機距離 (km)':<15} | {'目標點緯度':<12} | {'目標點經度':<12}")
print("-" * 60)
for pt in points[:10]:
    print(f"{pt.bearing:<12.1f} | {pt.distance_km:<15.3f} | {pt.coordinate.latitude:<12.6f} | {pt.coordinate.longitude:<12.6f}")
print("... (其餘省略)")
```

### 4. 進階：使用 Matplotlib 將結果視覺化（選用）
如果您有安裝 `matplotlib`，可以執行以下程式碼將船艦與產生的周圍點繪製在平面圖上。

```python
try:
    import matplotlib.pyplot as plt
    
    # 提取經緯度座標
    lats = [pt.coordinate.latitude for pt in points]
    lons = [pt.coordinate.longitude for pt in points]
    
    # 繪製圖表
    plt.figure(figsize=(8, 8))
    
    # 繪製周圍點 (散佈圖)
    plt.scatter(lons, lats, color='blue', alpha=0.6, label='Surrounding Points (5-15km)')
    
    # 繪製中心目標船艦
    plt.scatter([center.longitude], [center.latitude], color='red', marker='^', s=150, label='Target Ship')
    
    # 繪製連接線 (選用，輔助觀看方位)
    for pt in points:
        plt.plot([center.longitude, pt.coordinate.longitude], [center.latitude, pt.coordinate.latitude], 'gray', linestyle='--', alpha=0.3)
        
    plt.title("Target Ship & Surrounding Points Distribution", fontsize=14)
    plt.xlabel("Longitude (經度)", fontsize=12)
    plt.ylabel("Latitude (緯度)", fontsize=12)
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend(loc='upper right')
    plt.gca().set_aspect('equal', adjustable='box') # 保持經緯度比例一致
    plt.show()

except ImportError:
    print("請先安裝 matplotlib 套件以繪製視覺化圖表。")
    print("安裝指令: pip install matplotlib 或 uv add matplotlib")
```
